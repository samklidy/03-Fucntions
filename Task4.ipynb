```python
documents = [ {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'} ]

directories = { '1': ['2207 876234', '11-2'], '2': ['10006'], '3': [] }

#Функции

def create_list_docs(documents, num):
    """
    the function creates the list of 'numbers' and check if num is there
    """
    list_document = []  
    for i in documents:
        list_document.append(i['number'])
        
    if num not in list_document:
        result = print('Результат: Документ не найден в базе')
        return result
    
def create_str_shelfs(directories):
    """ 
    the function creates str of shelf numbers from dic directories
    """
    for key in directories.keys():
        e = (','.join(list(directories.keys())))
        return e
    
    
def commande_p(documents):
    """ 
    the function searches and prints name by document number 
    """
    num = str(input('Введите номер документа '))
    create_list_docs(documents, num)
    for i in documents:
        if i['number'] == num:
            a = i['name']
            result = print(f'Результат: Владелец документа: {a}')
            return result
        
def commande_s(documents, directories):
    """ 
    the function searches and prints shelf by document number and print
    """ 
    num = str(input('Введите номер документа '))
    create_list_docs(documents, num)
    for key, value in directories.items():
        if num in value:
            result = print(f'Результат: Документ хранится на полке: {key}')
            return result
        
def commande_l(documents, directories):
    """ 
    the function searches and print all info in documents and directories
    """  
    print('Результат: ')
    for i in documents:
        a = i['number']
        b = i['type']
        c = i['name']
        for key, value in directories.items():
            if i['number'] in value:
                d = key
        result = print(f'№: {a}, тип: {b}, владелец: {c}, полка хранения: {d}')
    return result

def commande_as(directories):
    """ 
    the function add a new shelf in directories
    """
    shelf = str(input('Введите номер полки '))
    if shelf in directories:
        print(f'Результат: Такая полка уже существует. Текущий перечень полок: {create_str_shelfs(directories)}.')
    else:
        directories[shelf] = []  
        print(f'Результат: Полка добавлена. Текущий перечень полок: {create_str_shelfs(directories)}.')

def commande_ds(directories):
    """ 
    the function deletes the shelf if it is empty
    """
    shelf = str(input('Введите номер полки '))
    if shelf not in directories.keys():
        print(f'Результат: Такой полки не существует. Текущий перечень полок: {create_str_shelfs(directories)}.')
    elif directories[shelf] == []:
        del(directories[shelf])
        print(f'Результат: Полка удалена. Текущий перечень полок: {create_str_shelfs(directories)}.')  
    else:
        print(f'Результат: На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {create_str_shelfs(directories)}.')
    
#Основной код

commande = str(input('Введите команду '))
while commande != 'q':
    if commande == 'p':
        commande_p(documents)
        commande = str(input('Введите команду '))
    elif commande == 's':
        commande_s(documents, directories)
        commande = str(input('Введите команду '))
    elif commande == 'l':
        commande_l(documents, directories)
        commande = str(input('Введите команду '))   
    elif commande == 'as':
        commande_as(directories)
        commande = str(input('Введите команду '))
    elif commande == 'ds':
        commande_ds(directories)
        commande = str(input('Введите команду '))
    else:
        commande = str(input('Введите команду '))
    
```
